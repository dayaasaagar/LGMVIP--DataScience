```python
import cv2
```

```python
#get the image location and the image file name.
img_loc="C:/Users/dayaaasaagar/"
filename = 'house.jpg'
```

```python
#read the image 
img = cv2.imread(img_loc+filename)
```

```python
print(img.shape)
```

```python
#show the image 
image=cv2.imshow('orginal image', img)
image
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#convert the image to gray scale
gray_image= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
plt.imshow(gray_image)
```

```python
#show the gray image
cv2.imshow('orginal Image',img)
cv2.imshow('new image', gray_image)
cv2.waitKey(0)
```

```python
#invert the image
invert_gray_theme= 255 - gray_image
plt.imshow(invert_gray_theme)
```

```python
#show the inverted gray image
cv2.imshow('orginal Image',img)
cv2.imshow('new image', invert_gray_theme)
cv2.waitKey(0)
```

```python
#blur the image by gaussian function
blur_im= cv2.GaussianBlur(invert_gray_theme, (21,21), 0)
plt.imshow(blur_im)
```

```python
#show the blurred image
cv2.imshow('orginal Image',img)
cv2.imshow('new image', blur_im)
cv2.waitKey(0)
```

```python
#invert the blurred image
inverted_blur_image= 255 - blur_im
plt.imshow(inverted_blur_image)
```

```python
cv2.imshow('orginal Image',img)
cv2.imshow('new image', inverted_blur_image)
cv2.waitKey(0)
```

```python
#create the pencil sketch 
sketch = cv2.divide(gray_image , inverted_blur_image, scale= 256.0)
plt.imshow(sketch)
```

```python
cv2.imshow('orginal Image',img)
cv2.imshow('new image', sketch)
cv2.waitKey(0)
```
